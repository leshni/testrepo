```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df_imi' DataFrame containing immigration data

# 1. Sort by Total Immigration and Select Bottom 5
df_imi.sort_values(['Total'], ascending=True, inplace=True)
df_bottom5 = df_imi.head(5)

# 2. Select Years and Transpose
years = list(map(str, range(1980, 2014)))
df_bottom5 = df_bottom5[years].transpose()

# 3. Convert Index to Integer for Plotting
df_bottom5.index = df_bottom5.index.map(int)

# 4. Create the Stacked Area Plot
df_bottom5.plot(kind='area',
                stacked=True,  # Stacked area plot
                alpha=0.55,     # Transparency
                figsize=(20, 10))

# 5. Add Labels and Title
plt.title('Immigration Trend of 5 Countries with Least Contribution')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.legend(loc='upper left')
plt.grid(True)

# 6. Display the Plot
plt.show()
```
