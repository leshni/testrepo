```python
#  list of years from 1980 - 2013 to start plotting the data
years = list(map(str, range(1980, 2014)))

years
```

```python
#Use sort values method to sort the df_imi Data Frame
#[Total] sorts the DF based on values in Total column

years = list(map(str, range(1980,2014)))

df_imi.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# select the top 5 entries
df_top_countries = df_imi.head()

# transpose the dataframe- swap thedata frame to get years to
# x axis and Imiigration numbers to y

df_top_countries = df_top_countries[years].transpose()

df_top_countries.head()
```

```python
df_imi.set_index('Country', inplace=True)

# Let's view the first five elements and see how the dataframe was changed
df_imi.head()
```

```python
df_imi.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# select the top 5 entries
df_top_countries = df_imi.head()

# transpose the dataframe
df_top_countries = df_top_countries[years].transpose()

df_top_countries.head()
```

```python
# Change the index values of df_top5 to type integer for plotting.
#This conversion is done to ensure compatibility
#with plotting functions that often require numerical index values.

# Define a custom color palette
colors = ['#9b59b6', '#e74c3c', '#34495e', '#2ecc71', '#f1c40f']
df_top_countries.index = df_top_countries.index.map(int)
#creating an area plot
df_top_countries.plot(kind='area',
             stacked=False,
             figsize=(20, 10),
             color=colors)  # pass a tuple (x, y) size
#Adding plot lables
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
#Display the plot
plt.show()
```

```python
# Change the index values of df_top5 to type integer for plotting.
#This conversion is done to ensure compatibility
#with plotting functions that often require numerical index values.
df_top_countries.index = df_top_countries.index.map(int)
#creating an area plot

# Define a new color palette (Seaborn's 'colorblind' palette)
colors = sns.color_palette("colorblind")

df_top_countries.plot(kind='area',
             stacked=False,
             figsize=(20, 10),
             color=colors)  # pass a tuple (x, y) size, and color list
#Adding plot lables
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
#Display the plot
plt.show()
```

```python
#Use the scripting layer to create a stacked area plot of the 5 countries that contributed the least to immigration to
#Canada from 1980 to 2013. Use a transparency value of 0.45.


# get the 5 countries with the least immigration
df_least5 = df_imi.tail(5)

# transpose the dataframe
df_least5 = df_least5[years].transpose()
df_least5.head()

 #change the index values of df_least5 to type integer for plotting
df_least5.index = df_least5.index.map(int)

#Alpha value gives the plot transparency
df_least5.plot(kind='area', alpha=0.45, figsize=(20, 10))

plt.title('Immigration Trend of 5 Countries with Least Contribution to Immigration')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```
